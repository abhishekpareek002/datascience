## Handling Missing values and deriving

## Standardizing the data 
- "revol_util" column although described as an object column, it has continous values. 
- So we need to standardize the data in this column

- "emp_length" --> { (< 1 year) is assumed as 0 and 10+ years is assumed as 10 } 

```python
loans.revol_util = pd.to_numeric(loans.revol_util.apply(lambda x : x.split('%')[0]))
```

```python
loans.int_rate = pd.to_numeric(loans.int_rate.apply(lambda x : x.split('%')[0]))
```

```python
loans.emp_length = pd.to_numeric(loans.emp_length.apply(lambda x: 0 if "<" in x else (x.split('+')[0] if "+" in x else x.split()[0])))
```

## Analysis of data

```python
# Method to Plot frequency

#Setting up background
sns.set(style="white")
def plt_frequency(x_var,data_df,x_label,title,hue_var=None,rot_flag=0,fig_size_x=15,fig_size_y=6):
    ncount = len(loans)
    dims = (fig_size_x,fig_size_y)
    plt.figure(figsize=dims)
    ax = sns.countplot(x=x_var, data=data_df,hue=hue_var,palette="Set3")
    plt.title(title)
    plt.xlabel(x_label)

    # Make twin axis
    ax2=ax.twinx()

    # Switch so count axis is on right, frequency on left
    ax2.yaxis.tick_left()
    ax.yaxis.tick_right()

    # Also switch the labels over
    ax.yaxis.set_label_position('right')
    ax2.yaxis.set_label_position('left')
    if(rot_flag==1):
        ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right", fontsize=12)
        plt.tight_layout()
    ax2.set_ylabel('Frequency [%]')

    for p in ax.patches:
        x=p.get_bbox().get_points()[:,0]
        y=p.get_bbox().get_points()[1,1]
        ax.annotate('{:.1f}%'.format(100.*y/ncount), (x.mean(), y), 
                ha='center', va='bottom') # set the alignment of the text

    # Need to turn the grid on ax2 off, otherwise the gridlines end up on top of the bars
    ax2.grid(None)
```

```python
plt_frequency("grade",loans,'Grade of Loan','Average distribution of loans across various Grades')
```

```python

plt_frequency("loan_status",loans,'Loan Status','Average overall Default rates')

```

```python

plt_frequency("issue_year",loans,'Loan Issue Year','Average overall distribution of loans in every year')
```

```python

plt_frequency("purpose",loans,'Purpose of Loan','Average distribution of loans across various categories',None,1)
```

```python
plt_frequency("emp_length",loans,'Employee Experience','Average distribution of loans across employee experience')
```
